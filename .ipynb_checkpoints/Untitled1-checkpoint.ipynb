```python
def canonicalize_smiles(df, smiles_col='smiles'):
    def get_canonical(smiles):
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            return Chem.MolToSmiles(mol, canonical=True)
        else:
            return smiles

    df[smiles_col] = df[smiles_col].apply(get_canonical)
    return df

df = canonicalize_smiles(df, smiles_col='smiles') 
duplicate_groups = df[df.duplicated(subset=['smiles'], keep=False)].groupby('smiles')

for smiles, group in duplicate_groups: 
    print("Canonical SMILES:", smiles)
    print(group)
    print()
```

```python
#utils
def smiles2morgan_raw(s, radius=2, nBits=2048):
    try:
        mol = Chem.MolFromSmiles(s)
        if mol is None:  
            return None
        fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=nBits)
        arr = np.zeros((1,), dtype=bool)
        DataStructs.ConvertToNumpyArray(fp, arr)
        arr = arr.astype(int)
        return arr
    except Exception as e:
        logging.warning(f"Error in smiles2morgan_raw for SMILES {s}: {str(e)}")
        return None

def generate_physchem_features(df):
    if 'SMILES' not in df.columns:
        raise ValueError("DataFrame must contain a 'SMILES' column")

    def calculate_properties(smiles):
        molecule = Chem.MolFromSmiles(smiles)
        if molecule is None:
            return [None] * 10

        logp = Crippen.MolLogP(molecule)
        mol_wt = Descriptors.MolWt(molecule)
        hbond_donor = Descriptors.NumHDonors(molecule)
        hbond_acceptor = Descriptors.NumHAcceptors(molecule)
        tpsa = rdMolDescriptors.CalcTPSA(molecule)
        num_rotatable_bonds = Descriptors.NumRotatableBonds(molecule)
        molecular_volume = Descriptors.ExactMolWt(molecule)
        num_aromatic_rings = Descriptors.NumAromaticRings(molecule)
        num_heteroatoms = Descriptors.NumHeteroatoms(molecule)
        num_aliphatic_rings = Descriptors.NumAliphaticRings(molecule)

        return [logp, mol_wt, hbond_donor, hbond_acceptor, tpsa, num_rotatable_bonds, molecular_volume, num_aromatic_rings, num_heteroatoms, num_aliphatic_rings]

    columns = ['LogP', 'MolecularWeight', 'HbondDonor', 'HbondAcceptor', 'TPSA', 'NumRotatableBonds', 'MolecularVolume', 'NumAromaticRings', 'NumHeteroatoms', 'NumAliphaticRings']
    df[columns] = df['SMILES'].apply(lambda x: pd.Series(calculate_properties(x)))

    return df

def smiles2maccs(s):
    mol = Chem.MolFromSmiles(s)
    if mol is None:
        return None
    return np.array(MACCSkeys.GenMACCSKeys(mol))

def smiles2rdkit2d(s):
    generator = rdNormalizedDescriptors.RDKit2DNormalized()
    features = np.array(generator.process(s)[1:])
    NaNs = np.isnan(features)
    features[NaNs] = 0
    return np.array(features)


def smiles2atompairs(s, nBits=2048):
    mol = Chem.MolFromSmiles(s)
    if mol is None:
        return None
    return np.array(rdMolDescriptors.GetHashedAtomPairFingerprintAsBitVect(mol, nBits=nBits))

def smiles2daylight(s):
    NumFinger = 2048
    mol = Chem.MolFromSmiles(s)
    bv = FingerprintMols.FingerprintMol(mol)
    temp = tuple(bv.GetOnBits())
    features = np.zeros((NumFinger, ))
    features[np.array(temp)] = 1
    return np.array(features)

def smiles2pubchem(s):
    features = calcPubChemFingerAll(s)
    return np.array(features)

def calcPubChemFingerAll(s):
    mol = Chem.MolFromSmiles(s)
    AllBits=[0]*881
    res1=list(calcPubChemFingerPart1(mol).ToBitString())
    for index, item in enumerate(res1[1:116]):
        if item == '1':
            AllBits[index] = 1
    for index2, item2 in enumerate(res1[116:734]):
        if item2 == '1':
            AllBits[index2+115+148] = 1
    res2=calcPubChemFingerPart2(mol)
    for index3, item3 in enumerate(res2):
        if item3==1:
            AllBits[index3+115]=1
    return np.array(AllBits)

def calcPubChemFingerPart1(mol, **kwargs):
  global PubchemKeys
  if PubchemKeys is None:
    PubchemKeys = [(None, 0)] * len(smartsPatts.keys())

    InitKeys(PubchemKeys, smartsPatts)
  ctor = kwargs.get('ctor', DataStructs.SparseBitVect)

  res = ctor(len(PubchemKeys) + 1)
  for i, (patt, count) in enumerate(PubchemKeys):
    if patt is not None:
      if count == 0:
        res[i + 1] = mol.HasSubstructMatch(patt)
      else:
        matches = mol.GetSubstructMatches(patt)
        if len(matches) > count:
          res[i + 1] = 1
  return res

def calcPubChemFingerPart2(mol):# 116-263
    bits=[0]*148
    bits=func_1(mol,bits)[1]
    bits=func_2(mol,bits)[1]
    bits=func_3(mol,bits)[1]
    bits=func_4(mol,bits)[1]
    bits=func_5(mol,bits)[1]
    bits=func_6(mol,bits)[1]
    bits=func_7(mol,bits)[1]
    bits=func_8(mol,bits)

    return bits
```

### Handle class imbalance


```python
print(selected_df['BBB'].value_counts())
```

```python
#ADASYN for class balance
majority_class = selected_df[selected_df['BBB'] == 0]
minority_class = selected_df[selected_df['BBB'] == 1]
undersampled_majority = majority_class.sample(minority_class.shape[0], random_state=42, replace=True)
balanced_df = pd.concat([undersampled_majority, minority_class])
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
balanced_df['BBB'].value_counts()
```

```python
#plot t-SNE
X = balanced_df.drop('BBB', axis=1)
y = balanced_df['BBB']
tsne = TSNE(n_components=2, random_state=42)
X_embedded = tsne.fit_transform(X)
plt.figure(figsize=(8, 6))
plt.scatter(X_embedded[y == 0, 0], X_embedded[y == 0, 1], label='0', alpha=0.5)
plt.scatter(X_embedded[y == 1, 0], X_embedded[y == 1, 1], label='1', alpha=0.5)
plt.title('t-SNE Visualization of Balanced Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.show()
```
