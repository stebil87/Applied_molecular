### AUC curve for baseline

```python
def plot_roc_curves(trained_models, test_dataset):
    fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 18)) 
    fig.suptitle('Receiver Operating Characteristic (ROC) Curves for Tested Models')
    
    keys = list(trained_models.keys())
    
    for i, ax in enumerate(axes.flat):
        if i < len(keys):
            models_dict = trained_models[keys[i]]
            for model_name, model in models_dict.items():
                if hasattr(model, "predict_proba"): 
                    y_score = model.predict_proba(test_dataset.X)[:, 1]
                    fpr, tpr, _ = roc_curve(test_dataset.y, y_score)
                    roc_auc = auc(fpr, tpr)
                    ax.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
                else:
                    print(f"Model {model_name} does not support predict_proba and will be skipped.")
            
            ax.plot([0, 1], [0, 1], 'k--', lw=2)
            ax.set_xlim([0.0, 1.0])
            ax.set_ylim([0.0, 1.05])
            ax.set_xlabel('False Positive Rate')
            ax.set_ylabel('True Positive Rate')
            ax.set_title(f'Models for {keys[i]}')
            ax.legend(loc="lower right")
        else:
            fig.delaxes(ax) 

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) 
    plt.show()

plot_roc_curves(trained_models, test_dataset)
```

```python
#test each test df subsplit wrt the right trained models, results and AUC 
def prepare_dataset(df):
    features = df.drop(columns=['BBB', 'smiles']).values
    targets = df['BBB'].values
    smiles = df['smiles'].values
    return dc.data.NumpyDataset(X=features, y=targets, ids=smiles)

dataset_sel = prepare_dataset(test_sel)
dataset_rus = prepare_dataset(test_rus_sel)
dataset_ros = prepare_dataset(test_ros_sel)

splitter = dc.splits.ScaffoldSplitter()
_, valid_dataset_sel, test_dataset_sel = splitter.train_valid_test_split(dataset_sel)
_, valid_dataset_rus, test_dataset_rus = splitter.train_valid_test_split(dataset_rus)
_, valid_dataset_ros, test_dataset_ros = splitter.train_valid_test_split(dataset_ros)

valid_datasets = {
    'Imbalanced selected df': valid_dataset_sel,
    'Randomly undersampled selected df': valid_dataset_rus,
    'Randomly oversampled selected df': valid_dataset_ros
}

test_datasets = {
    'Imbalanced selected df': test_dataset_sel,
    'Randomly undersampled selected df': test_dataset_rus,
    'Randomly oversampled selected df': test_dataset_ros
}

def test_models_and_compute_auc(trained_models_sel, datasets, set_name):
    results = {}
    roc_data = {}
    for df_name, models in trained_models_sel.items():
        dataset = datasets[df_name]
        features = dataset.X
        targets = dataset.y
        results[df_name] = {}
        roc_data[df_name] = {}
        for model_name, model in models.items():
            if hasattr(model, "predict_proba"):
                y_proba = model.predict_proba(features)[:, 1]
            else:
                y_proba = model.decision_function(features)
                y_proba = (y_proba - y_proba.min()) / (y_proba.max() - y_proba.min())  

            y_pred = model.predict(features)
            accuracy = accuracy_score(targets, y_pred)
            fpr, tpr, thresholds = roc_curve(targets, y_proba)
            roc_auc = auc(fpr, tpr)
            
            results[df_name][model_name] = accuracy
            roc_data[df_name][model_name] = (fpr, tpr, roc_auc)
    return results, roc_data

validation_scores, validation_roc = test_models_and_compute_auc(trained_models_sel, valid_datasets, "Validation Set")
test_scores, test_roc = test_models_and_compute_auc(trained_models_sel, test_datasets, "Test Set")

def print_scores(scores, set_name):
    print(f"\n{set_name} Results:")
    for df_name, model_results in scores.items():
        print(f"\n  {df_name}:")
        for model_name, accuracy in model_results.items():
            print(f"    {model_name}: Accuracy = {accuracy:.4f}")

def plot_roc_curves(roc_data, set_name):
    num_datasets = len(roc_data)
    fig, axes = plt.subplots(nrows=num_datasets, ncols=1, figsize=(10, 6 * num_datasets))
    fig.suptitle(f'ROC Curves for {set_name}')

    if num_datasets > 1:
        axes = axes.flatten()
    else:
        axes = [axes]  

    for ax, (df_name, models) in zip(axes, roc_data.items()):
        for model_name, (fpr, tpr, roc_auc) in models.items():
            ax.plot(fpr, tpr, label=f'{model_name} ({df_name}): AUC = {roc_auc:.2f}')
        ax.plot([0, 1], [0, 1], 'k--', lw=2)
        ax.set_xlim([0.0, 1.0])
        ax.set_ylim([0.0, 1.05])
        ax.set_xlabel('False Positive Rate')
        ax.set_ylabel('True Positive Rate')
        ax.set_title(f'Models for {df_name}')
        ax.legend(loc="lower right")
        ax.grid(True)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

print_scores(test_scores, "Test Set")
plot_roc_curves(test_roc, "Test Set")
```
